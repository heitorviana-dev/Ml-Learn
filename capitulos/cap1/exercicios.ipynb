## Exercícios

1 - O aprendizado de máquina é a ciência de como computadores aprendem baseado em uma experiência a serem melhores em uma tarefa baseado em uma medida de desempenho.

2 - Problemas complexos para abordagens tradicionais, Problemas que são extensas listas de regras, Descoberta de padrões(Mineração de dados) e Ajuste automático a novos dados.

3 - Um conjunto de dados que já possuem as soluções desejadas.

4 - Classificação e Regressão.

5 - Redução de dimensionalidade, Clusterização, Aprendizado de regras por associação e Detecção de anomalias/novidades.

6 - Um algoritmo de aprendizado por reforço, visto que essa situação é onde esse tipo de aprendizado brilha.

7 - Eu utilizaria um algoritmo de Clusterização.

8 - A tarefa de detecção de spam é um exemplo clássico de aprendizado supervisionado.

9 - Um sistema que aprende em incrementos.

10 - Aprendizado utilizado em situações cujos dados não cabem na máquina, assim divide-se os dados em mini batches, utilizando técnicas de aprendizado online para aprender a partir desses pequenos lotes.

11 - Aprendizado baseado em instâncias.

12 - Um parâmetro de modelo é definido a partir dos dados utilizados no treino do modelo, ele define o ajuste do modelo aos dados, enquanto o hiperparâmetro é definido pelo algoritmo de aprendizado, não podendo ser alterado durante o treino.

13 - Eles servem para realizar previsões baseado em um treinamento prêvio com dados. Para ajustar os parâmetros do modelo, isto é, para treinar o modelo ele busca minimizar uma função de custo, ou seja, minimizar o erro de generalização.

14 - Dados de baixa qualidade(Entra lixo, sai lixo), Overfitting, Underfitting, pouca representatividade dos dados, Quantidade insuficiente de dados e características irrelevantes.

15 - Está acontecendo o Overfitting, isto é, o sobreajuste nos dados de treinamento, para resolvé-lo podemos simplificar o modelo aplicando a regularização, coletar novos dados e limpar os dados atuais.

16 - Um conjunto de testes serve para prever o erro de generalização que ocorrerá em produção. Eu utilizaria para conseguir prever como o meu modelo irá sair com os dados recebidos em produção.

17 - Um conjunto de validação serve para adicionar maior precisão ao modelo, a partir dele conseguimos ajustar os hiperparâmetros com maior precisão, evitando overfitting dos dados.

18 - O train dev set é um divisão do conjunto de treinamento que é usado em situações onde há suspeitas de incompatibilidade dos dados, a partir dele podemos verificar se o erro se deu devido ao overfitting com os dados de treinamento ou a partir da incompatibilidade dos dados. Para usá-lo, primeiro treina-se o modelo com o conjunto de treinamento - train dev set, depois testa com o train dev set, se ele tiver um bom desempenho com o train dev, então ele não está sofrendo de overfitting, se ele tiver um desempenho ruim com o conjunto de validação, então há grandes chances de ele estar sofrendo com incompatibilidade dos dados.

19 - Pode ocorrer o vício nos dados, isto é, overfitting nos dados de treinamento.
