# Capítulo 1 - O Cenário do Aprendizado de Máquina

## O que é Machine Learning?

- Existem duas definições que ao meu ver podem ser consideradas as principais, uma mais generalista definida por Arthur Samuel em 1959 e uma mais técnica e moderna definida por Tom Mitchell em 1997.

- Segundo Arthur Samuel, o aprendizado de máquina é a habilidade do computador de aprender sem ser explicitamente programados.

- Segundo Tom Mitchell, um computador aprende se a partir de uma experiência E o modelo consegue desempenhar melhor na tarefa T, baseado na métrica de desempenho P. Por exemplo, consideremos E um dataset de imóveis, a tarefa T prever o valor dos imóveis e a métrica P como sendo a RMSE (Root Mean Square Error), então um modelo aprende se a partir do dataset ele conseguiu prever melhor o valor dos imóveis, sendo metrificado pela métrica RMSE.

- Chamados os dados usados no treinamento de **Conjunto de treinamento** e um único dado é chamado de **instância de treino** ou **amostra**.

## Porque usar Machine Learning?

- Agora um dos pontos mais importantes, porque raios eu iria estudar machine learning? Imagine comigo, como você criaria um filtro de spam utilizando a abordagem tradicional da programação, primeiro analisaríamos o problema e os dados, começaríamos a enxergar padrões de palavras que indiquem que o email é um spam, criaríamos regras para verificar se o email contém essas combinações de palavras e o disponibilizaríamos para fazer testes. Porém vamos analisar alguns problemas com essa abordagem

- Primeiro, concorda comigo que muitos padrões de spam seriam perdidos durante a análise? Você poderia encontrar que a combinação de palavras só pra você pode indicar um spam, porém você poderia deixar passar outros padrões que na sua visão não indicariam o spam. Segundo, o sistema seria uma extensa lista de regras que a medida que novos tipos de spam surgissem maior e mais difícil de manter o sistema ficaria, sem contar que para cada novo padrão a lista deveria ser mudada, logo teríamos um trabalho praticamente interminável.

- Mas e se utilizássemos um modelo de ML que passados os emails com spam e sem spam (ham), detectasse padrões automaticamente, preveria se o email é spam ou não, se além disso também automatizassemos o treinamento dele permitindo que ele aprendesse novos padrões sem que nós intervíssemos? Vamos concordar que parece bem melhor, mais simples, mais fácil de manter e mais preciso.

- Então, o Aprendizado de Máquina pode resolver problemas que para a abordagem tradicional seriam muito complexos ou difíceis de serem mantidos, como o caso do filtro de spam, porém, além disso, o aprendizado de máquina é utilizado para descobrir padrões implícitos em grandes volumes de dados, a chamada mineração de dados, mas não só em grandes volumes mas em outros datasets.

- Logo, o AM é muito utilizado para:
    - Resolver problemas que utilizando a abordagem tradicional seriam extensas listas de regras.
    - Resolver problemas que utilizando a abordagem tradicional seriam complexos ou ainda não possuem algoritmos conhecidos.
    - Por ser altamente adaptável ao automatizar o pipeline de dados.
    - Para descobrir padrões implícitos em grandes volumes de dados.
