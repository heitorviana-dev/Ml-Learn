# Capítulo 1 - O Cenário do Aprendizado de Máquina

## O que é Machine Learning?

- Existem duas definições que ao meu ver podem ser consideradas as principais, uma mais generalista definida por Arthur Samuel em 1959 e uma mais técnica e moderna definida por Tom Mitchell em 1997.

- Segundo Arthur Samuel, o aprendizado de máquina é a habilidade do computador de aprender sem ser explicitamente programados.

- Segundo Tom Mitchell, um computador aprende se a partir de uma experiência E o modelo consegue desempenhar melhor na tarefa T, baseado na métrica de desempenho P. Por exemplo, consideremos E um dataset de imóveis, a tarefa T prever o valor dos imóveis e a métrica P como sendo a RMSE (Root Mean Square Error), então um modelo aprende se a partir do dataset ele conseguiu prever melhor o valor dos imóveis, sendo metrificado pela métrica RMSE.

- Chamados os dados usados no treinamento de **Conjunto de treinamento** e um único dado é chamado de **instância de treino** ou **amostra**.

## Porque usar Machine Learning?

- Agora um dos pontos mais importantes, porque raios eu iria estudar machine learning? Imagine comigo, como você criaria um filtro de spam utilizando a abordagem tradicional da programação, primeiro analisaríamos o problema e os dados, começaríamos a enxergar padrões de palavras que indiquem que o email é um spam, criaríamos regras para verificar se o email contém essas combinações de palavras e o disponibilizaríamos para fazer testes. Porém vamos analisar alguns problemas com essa abordagem

- Primeiro, concorda comigo que muitos padrões de spam seriam perdidos durante a análise? Você poderia encontrar que a combinação de palavras só pra você pode indicar um spam, porém você poderia deixar passar outros padrões que na sua visão não indicariam o spam. Segundo, o sistema seria uma extensa lista de regras que a medida que novos tipos de spam surgissem maior e mais difícil de manter o sistema ficaria, sem contar que para cada novo padrão a lista deveria ser mudada, logo teríamos um trabalho praticamente interminável.

- Mas e se utilizássemos um modelo de ML que passados os emails com spam e sem spam (ham), detectasse padrões automaticamente, preveria se o email é spam ou não, se além disso também automatizassemos o treinamento dele permitindo que ele aprendesse novos padrões sem que nós intervíssemos? Vamos concordar que parece bem melhor, mais simples, mais fácil de manter e mais preciso.

- Então, o Aprendizado de Máquina pode resolver problemas que para a abordagem tradicional seriam muito complexos ou difíceis de serem mantidos, como o caso do filtro de spam, porém, além disso, o aprendizado de máquina é utilizado para descobrir padrões implícitos em grandes volumes de dados, a chamada mineração de dados, mas não só em grandes volumes mas em outros datasets.

- Logo, o AM é muito utilizado para:
    - Resolver problemas que utilizando a abordagem tradicional seriam extensas listas de regras.
    - Resolver problemas que utilizando a abordagem tradicional seriam complexos ou ainda não possuem algoritmos conhecidos.
    - Por ser altamente adaptável ao automatizar o pipeline de dados.
    - Para descobrir padrões implícitos em grandes volumes de dados.

## Tipos de sistemas de Machine Learning:

- Para classificarmos os vários modelos de machine learning que existem vamos utilizar as seguintes classificações:

    - Aprendizado Supervisionado x Não Supervisionado x Semissupervisionado x Por Reforço

    - Aprendizado em Batch x Online

    - Aprendizado baseado em instâncias x em modelo

### Aprendizado Supervisionado:

- O **aprendizado supervisionado** aborda modelos cujo conjunto de treinamento já está rotulado, isto é, já possui a solução esperada, por exemplo, suponha que queremos prever o valor de um imóvel, no nosso conjunto de treino o valor real deu 130000, já o modelo retornou 125000, a partir dessa comparação podemos metrificar o nosso modelo, tornando o melhor para a tarefa.

- Dentre os principais tipos de tarefas realizadas temos as de **classificação**, a qual foca em classificar novos em dados em classes, e as de **regressão**, que visam prever um valor numérico. Alguns algoritmos de regressão são utilizados em tarefas de classificação, como a regressão logística a qual foca em prever um valor entre 0 e 1, logo pode prever a probabilidade de um novo dado ser pertencente a uma classe x, e vice e versa.

- Alguns dos principais algoritmos de aprendizado supervisionado são:

    - Regressão Linear
    
    - Regressão Logística

    - Árvores de Decisão e Random Forest

    - Redes Neurais (Algumas)

    - Máquina de vetores de suporte (SVM)

    - K-ésimo vizinho mais próximo

### Aprendizado não supervisionado:

- O **Aprendizado não supervisionado**, como esperado, aborda modelos cujo conjunto de treinamento não estão rotulados. Esse tipo de aprendizado foca em resolver problemas em que queremos classificar ou descobrir padrões no nosso conjunto de dados, porém não temos ideia de como classificar.

- Por exemplo, imagine que você trabalha no setor de dados de um supermercado e o seu chefe te deu um conjunto de dados e quer que você analise eles, porém você olha pros dados e não encontra nenhuma variável alvo(target) para prever e nem classes para classificar, utilizando algoritmos de aprendizado não supervisionado, como os de clusterização, você pode descobrir grupos de clientes e assim conseguir criar estratégias de marketing melhores ou mudar os produtos de lugar.

- Dentro os algoritmos de Aprendizado não supervisionado temos os de **clusterização**, os quais recebem dados brutos não rotulados e buscam criar clusters, isso é grupos, em resumo classes.

- A **redução de dimensionalidade** é outro tipo de algoritmo de aprendizado não supervisionado muito importante o qual busca características correlacionadas e faz o merge delas em uma única variável, esse merge é chamado de **extração de características**. Esse algoritmo é muito utilizado para reduzir a dimensão do conjunto de treinamento, diminuindo o tamanho do dataset e assim o consumo de espaço, e fornecé-lo para um outro algoritmo de ML, podendo melhorar o seu desempenho.

- A **detecção de anomalias** e a **detecção de novidades** são tarefas muito importantes que utilizam algoritmos de aprendizado não supervisionado, eles treinam o modelo com instâncias normais e, quando um novo dado é inserido, eles buscam detectar se são anomalias ou novidades. Esses algoritmos são usados para remover outliers do conjunto de treinamento, repassando os dados para um outro algoritmo de ML.

- Por fim, temos o **Aprendizado de regras por associação** que busca descobrir relações interessantes entre os atributos de um conjunto de dados.

### Aprendizando semissupervisionado:

- Em muitos casos da vida real, os dados que você terá não estarão totalmente rotulados, muitos estarão parcialmente ou não rotulados, mas porque? Pois rotular dados é algo que consome tempo e dinheiro. Para lidar com esses tipos de dados, podemos utilizar algoritmos de **Aprendizado semissupervisionado**, os quais geralmente mesclam algoritmos de aprendizado supervisionado e não supervisionado.

### Aprendizado por Reforço:

- Imagine que temos um agente, esse agente precisa realizar uma tarefa T, para isso ele utiliza uma estratégia, chamada de *Política*, a partir dessa ele realiza ações, as quais se o aproximarem do objetivo, o agente será recompensado, e se não, ele será punido. O **Aprendizado por Reforço** é isso, ele treina agentes a partir de um sistema de recompensas para suas ações, sendo muito utilizado em jogos, por exemplo.

### Aprendizado em Batch:

- Distoando um pouco dos aprendizados discutidos acima, vamos agora classificar os modelos em como eles são treinados. No **Aprendizado em batch**, os modelos são treinados de uma vez com um conjunto de dados e, assim que implementados, para atualizar com novos dados será preciso treiná-los novamente com os novos e antigos dados. Logo, os modelos são treinados do zero e se quisermos atualizar com novos dados, devemos treiná-lo do zero novamente.

- Esse tipo de aprendizado é útil para tarefas que não necessitam de dados em tempo real, já que utilizar esse tipo de aprendizado para em modelos com treinamentos automatizados custaria uma fortuna.
