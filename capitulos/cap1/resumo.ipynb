# Capítulo 1 - O Cenário do Aprendizado de Máquina

## O que é Machine Learning?

- Existem duas definições que ao meu ver podem ser consideradas as principais, uma mais generalista definida por Arthur Samuel em 1959 e uma mais técnica e moderna definida por Tom Mitchell em 1997.

- Segundo Arthur Samuel, o aprendizado de máquina é a habilidade do computador de aprender sem ser explicitamente programados.

- Segundo Tom Mitchell, um computador aprende se a partir de uma experiência E o modelo consegue desempenhar melhor na tarefa T, baseado na métrica de desempenho P. Por exemplo, consideremos E um dataset de imóveis, a tarefa T prever o valor dos imóveis e a métrica P como sendo a RMSE (Root Mean Square Error), então um modelo aprende se a partir do dataset ele conseguiu prever melhor o valor dos imóveis, sendo metrificado pela métrica RMSE.

- Chamados os dados usados no treinamento de **Conjunto de treinamento** e um único dado é chamado de **instância de treino** ou **amostra**.

## Porque usar Machine Learning?

- Agora um dos pontos mais importantes, porque raios eu iria estudar machine learning? Imagine comigo, como você criaria um filtro de spam utilizando a abordagem tradicional da programação, primeiro analisaríamos o problema e os dados, começaríamos a enxergar padrões de palavras que indiquem que o email é um spam, criaríamos regras para verificar se o email contém essas combinações de palavras e o disponibilizaríamos para fazer testes. Porém vamos analisar alguns problemas com essa abordagem

- Primeiro, concorda comigo que muitos padrões de spam seriam perdidos durante a análise? Você poderia encontrar que a combinação de palavras só pra você pode indicar um spam, porém você poderia deixar passar outros padrões que na sua visão não indicariam o spam. Segundo, o sistema seria uma extensa lista de regras que a medida que novos tipos de spam surgissem maior e mais difícil de manter o sistema ficaria, sem contar que para cada novo padrão a lista deveria ser mudada, logo teríamos um trabalho praticamente interminável.

- Mas e se utilizássemos um modelo de ML que passados os emails com spam e sem spam (ham), detectasse padrões automaticamente, preveria se o email é spam ou não, se além disso também automatizassemos o treinamento dele permitindo que ele aprendesse novos padrões sem que nós intervíssemos? Vamos concordar que parece bem melhor, mais simples, mais fácil de manter e mais preciso.

- Então, o Aprendizado de Máquina pode resolver problemas que para a abordagem tradicional seriam muito complexos ou difíceis de serem mantidos, como o caso do filtro de spam, porém, além disso, o aprendizado de máquina é utilizado para descobrir padrões implícitos em grandes volumes de dados, a chamada mineração de dados, mas não só em grandes volumes mas em outros datasets.

- Logo, o AM é muito utilizado para:
    - Resolver problemas que utilizando a abordagem tradicional seriam extensas listas de regras.
    - Resolver problemas que utilizando a abordagem tradicional seriam complexos ou ainda não possuem algoritmos conhecidos.
    - Por ser altamente adaptável ao automatizar o pipeline de dados.
    - Para descobrir padrões implícitos em grandes volumes de dados.

## Tipos de sistemas de Machine Learning:

- Para classificarmos os vários modelos de machine learning que existem vamos utilizar as seguintes classificações:

    - Aprendizado Supervisionado x Não Supervisionado x Semissupervisionado x Por Reforço

    - Aprendizado em Batch x Online

    - Aprendizado baseado em instâncias x em modelo

### Aprendizado Supervisionado:

- O **aprendizado supervisionado** aborda modelos cujo conjunto de treinamento já está rotulado, isto é, já possui a solução esperada, por exemplo, suponha que queremos prever o valor de um imóvel, no nosso conjunto de treino o valor real deu 130000, já o modelo retornou 125000, a partir dessa comparação podemos metrificar o nosso modelo, tornando o melhor para a tarefa.

- Dentre os principais tipos de tarefas realizadas temos as de **classificação**, a qual foca em classificar novos em dados em classes, e as de **regressão**, que visam prever um valor numérico. Alguns algoritmos de regressão são utilizados em tarefas de classificação, como a regressão logística a qual foca em prever um valor entre 0 e 1, logo pode prever a probabilidade de um novo dado ser pertencente a uma classe x, e vice e versa.

- Alguns dos principais algoritmos de aprendizado supervisionado são:

    - Regressão Linear
    
    - Regressão Logística

    - Árvores de Decisão e Random Forest

    - Redes Neurais (Algumas)

    - Máquina de vetores de suporte (SVM)

    - K-ésimo vizinho mais próximo

### Aprendizado não supervisionado:

- O **Aprendizado não supervisionado**, como esperado, aborda modelos cujo conjunto de treinamento não estão rotulados. Esse tipo de aprendizado foca em resolver problemas em que queremos classificar ou descobrir padrões no nosso conjunto de dados, porém não temos ideia de como classificar.

- Por exemplo, imagine que você trabalha no setor de dados de um supermercado e o seu chefe te deu um conjunto de dados e quer que você analise eles, porém você olha pros dados e não encontra nenhuma variável alvo(target) para prever e nem classes para classificar, utilizando algoritmos de aprendizado não supervisionado, como os de clusterização, você pode descobrir grupos de clientes e assim conseguir criar estratégias de marketing melhores ou mudar os produtos de lugar.

- Dentro os algoritmos de Aprendizado não supervisionado temos os de **clusterização**, os quais recebem dados brutos não rotulados e buscam criar clusters, isso é grupos, em resumo classes.

- A **redução de dimensionalidade** é outro tipo de algoritmo de aprendizado não supervisionado muito importante o qual busca características correlacionadas e faz o merge delas em uma única variável, esse merge é chamado de **extração de características**. Esse algoritmo é muito utilizado para reduzir a dimensão do conjunto de treinamento, diminuindo o tamanho do dataset e assim o consumo de espaço, e fornecé-lo para um outro algoritmo de ML, podendo melhorar o seu desempenho.

- A **detecção de anomalias** e a **detecção de novidades** são tarefas muito importantes que utilizam algoritmos de aprendizado não supervisionado, eles treinam o modelo com instâncias normais e, quando um novo dado é inserido, eles buscam detectar se são anomalias ou novidades. Esses algoritmos são usados para remover outliers do conjunto de treinamento, repassando os dados para um outro algoritmo de ML.

- Por fim, temos o **Aprendizado de regras por associação** que busca descobrir relações interessantes entre os atributos de um conjunto de dados.

### Aprendizando semissupervisionado:

- Em muitos casos da vida real, os dados que você terá não estarão totalmente rotulados, muitos estarão parcialmente ou não rotulados, mas porque? Pois rotular dados é algo que consome tempo e dinheiro. Para lidar com esses tipos de dados, podemos utilizar algoritmos de **Aprendizado semissupervisionado**, os quais geralmente mesclam algoritmos de aprendizado supervisionado e não supervisionado.

### Aprendizado por Reforço:

- Imagine que temos um agente, esse agente precisa realizar uma tarefa T, para isso ele utiliza uma estratégia, chamada de *Política*, a partir dessa ele realiza ações, as quais se o aproximarem do objetivo, o agente será recompensado, e se não, ele será punido. O **Aprendizado por Reforço** é isso, ele treina agentes a partir de um sistema de recompensas para suas ações, sendo muito utilizado em jogos, por exemplo.

### Aprendizado em Batch:

- Distoando um pouco dos aprendizados discutidos acima, vamos agora classificar os modelos em como eles são treinados. No **Aprendizado em batch**, os modelos são treinados de uma vez com um conjunto de dados e, assim que implementados, para atualizar com novos dados será preciso treiná-los novamente com os novos e antigos dados. Logo, os modelos são treinados do zero e se quisermos atualizar com novos dados, devemos treiná-lo do zero novamente.

- Esse tipo de aprendizado é útil para tarefas que não necessitam de dados em tempo real, já que utilizar esse tipo de aprendizado para em modelos com treinamentos automatizados custaria uma fortuna.

### Aprendizado Online (Incremental):

- O **Aprendizado Online** ou **Aprendizado Incremental** visa dividir os dados de treina em pequenos lotes chamados de **mini batches**, o treinamento do modelo é feito aos poucos de modo a ocupar menos espaço.

- Os dois principais problemas que o Aprendizado Online resolve é a automatização do treinamento do modelo com novos dados, em que o modelo pode ser treinado constantemente com novos dados inseridos em pequenos lotes, desse modo não é necessário parar o modelo e treiná-lo do zero novamente como acontecia no Aprendizado em Batch.

- O segundo problema seria a falta de recursos computacionais, com o Aprendizado Online você não precisa subir o dataset de uma vez para a sua máquina, você pode treinar o seu modelo dividindo esse dataset em pequenas partes, de modo que dado que foi treinado por uma das partes você pode descartá-la, assim você treina o seu modelo com todos os dados, porém não sobe o dataset inteiro, esse tipo de aprendizado também pode ser chamado de **Aprendizado Out-of-Core**.

- Como estamos falando de incrementar o nosso modelo com novos dados, vale a pena falarmos sobre um parâmetro chamado de **taxa de aprendizado**(α). Esse parâmetro indica o quão rápido o modelo irá se adaptar à novos dados, quanto maior ele for mais rápido será, porém com um preço, ele irá se esquecer mais rápido dos dados antigos, desse modo ele ficaria restrito, ou melhor, enviesado com os novos dados, quanto menor ele for mais inerte ele será a novos dados, porém a chance do problema de viês à novos dados será diminuído consideravelmente.

### Aprendizado baseado em instâncias:

- Classificamos os algoritmos baseado em como o dataset está rotulado e em como os dados são inseridos, agora vamos classificar em relação à como o treinamento do modelo ocorre. 

- Nos algoritmos de **Aprendizado baseado em instâncias** os exemplos de treino são armazenados em sua totalidade na máquina, de modo que o algoritmo irá classificar os novos dados baseado em uma **medida de similaridade** com os exemplos de treino. Logo, nós não treinamos um modelo, nós literalmente armazenamos todo o dataset e baseado nos exemplos contidos nele classificamos os novos.

- O principal algoritmo desse tipo de aprendizado é o **K-Nearest Neighbors(KNN ou K-ésimo vizinho mais próximo)**, ele classifica os novos dados baseado na distância com os exemplos de treino, imagine que temos um dataset de Satisfação x Pib, em que queremos prever o nível de satisfação de um país baseado no seu Pib:


baseado na distância com os outros países, o sistema irá prever o nível de satisfação do novo país.

- O principal problema desse tipo de aprendizado é armazenar o dataset inteiro na sua máquina, porém mesmo com esse problema, o algoritmo KNN é um dos mais utilizados em ML por sua simplicidade.

### Aprendizado baseado em modelo:

- Agora sim vamos tratar sobre o treinamento de um modelo, no **Aprendizado baseado em modelo** o conjunto de treinamento é utilizado para treinar o modelo, e depois disso é descartado caso não seja necessário o rollback dos dados(retornar a como o modelo estava antes), mas o que é treinar um modelo? Como que um modelo realmente aprende?

- Imagine que temos o mesmo dataset de Satisfação x Pib, em que prevemos a satisfação baseado exclusivamente no Pib, e antes de escolhermos o modelo, vamos analisar os dados, para isso vamos plotar(exibir) um gráfico de dispersão(scatter plot):



veja que ao olhar para o gráfico podemos traçar uma linha reta para representar os dados, para isso vamos utilizar uma função linear y = a*x + b, em que o parâmetro a é a inclinação da reta e o parâmetro b representa onde a reta corta o eixo y. Certo, eu sei que x é o Pib e y é a satisfação, porém como eu defino os parâmetros? É ai meu pequeno padawan que o treinamento entra.

- *Treinar um modelo significa ajustar os parâmetros para o modelo refletir melhor aos dados*, logo o treinamento serve para ajustar o modelo aos dados. O treinamento ocorre da seguinte maneira, escolhemos uma medida de desempenho podendo ser uma **função de utilidade**, a qual mede o quão bom o modelo é, ou uma **função de custo**, que mede o quão ruim o modelo é, no caso de modelos de função linear, isto é, regressão linear, utilizamos mais as funções de custo. Escolhido a função agora nós vamos querer **minimizar o erro obtido**, então no começo escolhemos parâmetros aleatórios e vamos tentando ao máximo diminuir o erro obtido na função de custo, não vamos entrar em detalhes agora em como que os parâmetros são escolhidoa para minimizar o erro, mas fique com essa ideia geral de minimização do erro. Feito o treinamento, agora podemos prever os novos dados, passo chamado de **inferência**.

- Então o processo de criar um projeto de ML consiste, de maneira bem geral, nos seguintes quatro passos:
    - Estudar o problema.
    - Selecionar o modelo.
    - Treinar o modelo.
    - Inferir novos resultados.

- Abordamos então algumas classificações utilizadas para classificar os diversos algoritmos de ML e como funciona o fluxo de um projeto de ML. Agora vamos tratar sobre os principais desafios de ML.

## Principais desafios de Machine Learning:

- Na vida real, treinar um modelo de ML, na maioria das vezes, não é um mar de rosas. Quando falamos de desafios podemos falar em relação aos dados e em relação aos algoritmos, vamos começar tratando dos dados.

### Quantidade insuficiente de dados:

- Quando vamos treinar um modelo precisamos, é claro, de uma quantidade grande de dados, geralmente em pequenos e médios projetos utilizados centenas ou milhares de dados para treinar o nosso modelo, porém em grandes projetos como visualização de imagem ou LLMs(Large Language Model) são utilizados milhões de dados de treino. 

- Treinar o modelo com poucos dados faz com que o mesmo sofra de problema de ruído, isto é, viês, pois imagine comigo nós estamos tentando prever valores para uma população a partir de uma amostra, e se essa amostra for pequena ela não irá conseguir representar a população.

### Dados pouco representativos:

- Além da quantidade insuficiente, precisamos entender que os dados precisam representar uma população, então o mesmo não deve possuir viêses como nós humanos temos na maioria das vezes, porém, assim como nós, os dados terão esses viêses, e se não tratarmos eles, isso acabará gerando padrões que não representam a realidade e assim o modelo irá desempenhar de maneira não satisfatória.

### Dados de baixa qualidade:

- Os cientistas de dados passam um tempo significativo fazendo o que chamamos de **limpeza dos dados**, uma parte da preparação dos dados, a partir dela podemos lidar com outliers(anomalias), dados faltantes, dentre outros, mas porque fazer isso?

- Quando treinamos o nosso modelo com dados de baixa qualidade ele irá aprender padrões que não representam a vida real, não retornando dados precisos e assim afetando a tomada de decisões.

### Características irrelevantes:

- Além da limpeza dos dados, outra parte muito importante da preparação dos dados é a redução de características irrelevantes, onde é feita a **extração de características** discutida anteriormente em redução de dimensionalidade. Chamamos esse processo de **Engenharia de Features**.

- Treinar um modelo com várias características que não afetam o resultado pode ocasionar problemas de desempenho e descoberta de padrões irrelevantes, por exemplo, imagine o mesmo dataset de Satisfação x Pib, imagine que adicionassemos uma variável nome do país, concorda comigo que o modelo poderia descobrir padrões que não representam a realidade, como países com W tem altos níveis de satisfação.

- Abordamos sobre problemas com dados, agora vamos falar sobre problemas no algoritmo.

### Overfitting (Sobreajuste dos dados):

- O **Overfitting** ocorre quando o modelo se ajusta muito bem aos dados de treino, porém quando o colocamos para inferir novos dados o mesmo não tem capacidade de generalizar, gerando um desempenho ruim.

- Para resolver isso temos três alternativas:

    - Simplificar o modelo.
    - Coletar novos dados.
    - Limpar os dados.

- Já falamos da limpeza, vamos abordar sobre a simplificação do modelo, como podemos fazer isso? A simplificação é feita a partir de um processo chamado de **regularização**, em que os **hiperparâmetros**, isto é, parâmetros do algoritmo que são definidos antes do treinamento e não podem ser alterados, são ajustados. Não iremos entrar a fundo em como isso acontece, porém essa é uma parte muito importante e que será abordada no próximo capítulo.

### Underfitting (Sobajuste dos dados):

- O **Underfitting** é o contrário do Overfitting, ele ocorre quando o modelo é simples demais e não consegue nem se ajustar aos dados. 

- Para resolver isso temos três opções:

    - Aplicar um modelo mais complexo.
    - Adicionar mais características relevantes a partir da engenharia de features.
    - Remover algumas restrições do modelo a partir do ajuste dos hiperparâmetros(Regularização).

- Falamos sobre os tipos de aprendizado, o fluxo de um projeto de ML e os principais desafios envolvendo dados e algoritmos, porém dado que fizemos todo o fluxo do projeto, vamos apenas torcer para que o modelo consiga generalizar bem? Claro que não, nós devemos avaliar e ajustar o nosso modelo seguindo algumas práticas que iremos abordar agora.

## Teste e Validação:

### Conjunto de teste e treino:

- Uma prática comum na ciência de dados é separar o conjunto de dados em duas partes, o **Conjunto de Treino** e o **Conjunto de Teste**, geralmente utilizando a proporção de 80 para 20, porém isso depende da quantidade de dados e seu objetivo com eles, mas onde cada conjunto será utilizado?

- O conjunto de treino é utilizado para fazer o treinamento do nosso modelo, isto é, o ajuste dos parâmetros, como discutido anteriormente, já o conjunto de teste é utilizado para fazermos predições com dados que possuímos a solução, dessa maneira podemos comparar o valor previsto com o valor real e assim obter o **erro de generalização**, sendo o nosso objetivo minimizar esse erro.

### Conjunto de validação - Método Holdout:

- Imagine a situação, você possui um conjunto de dados e está em dúvida se escolhe o modelo A ou o modelo B, como resolver a situação? Uma solução seria testar os modelos A e B e comparar o erro de generalização, aquele que tiver o menor erro será utilizado.

- Certo, suponha que o modelo A foi o melhor, mas agora você quer ajustar os hiperparâmetros para mitigar o Overfitting, como fazer? Escolher valores aleatórios? Uma solução seria pegar vários modelos com diferentes valores de hiperparâmetros e testá-los, aquele que apresentar o menor erro de generalização é escolhido.

- Escolhido o modelo você obteve apenas 5% de erro de generalização, assim você sobe ele em produção e o disponibiliza para os seus usuários. Porém ao fazer o monitoramento do modelo com os dados obtidos em produção você percebeu que o erro de generalização deu 15%, o que deu de errado?

- O que aconteceu foi o seguinte, o seu modelo ficou viciado, isto é, ele ajustou os hiperparâmetros muito bem para os dados de treino, e como ele foi treinado e testado com os mesmos dados, isso acabou o deixando preso a esses dados, mas como eu posso mitigar esse vício?

- Para fazer isso podemos usar o **Método Holdout de validação cruzada**, o qual não tem validação cruzada nenhuma, mas como o autor descreveu assim vamos manter a tipologia, esse método divide o conjunto de dados na seguinte divisão estática, conjunto de treino, conjunto de teste e **conjunto de validação**, e agora o processo de treino e teste será realizado da seguinte maneira:

    - Primeiro vamos treinar os nossos modelos com diferentes valores de hiperparâmetros com um conjunto de treino limitado (conjunto de treino - conjunto de validação) e testá-lo com o conjunto de validação.

    - Escolhido o modelo que melhor apresentou o menor erro de generalização agora vamos treiná-lo com o conjunto de treino e testá-lo com o conjunto de teste, e assim obtemos um modelo mais preciso que utilizando apenas os conjuntos de treino e teste.

- Esse método funciona na maioria das situação, porém existe algumas que ele acaba falhando, como quando o conjunto de validação é muito pequeno e, por isso, o modelo com o hiperparâmetro ajustado pode ter sido escolhido ao acaso, se o conjunto de validação for muito grande, o modelo acaba sendo treinado mal, já que o conjunto de treino, que serve como a base do modelo, não é satisfatório, então para resolver isso podemos usar um outro método.

### Conjunto de validação - Método K-Fold de validação cruzada:

- Agora sim teremos um método onde realmente ocorre a validação cruzada, nesse método nós dividimos o conjunto de treino em "dobras", que nada mais são que divisões, funcionando da seguinte maneira:



- Cada modelo agora será treinado e testado com um conjunto de treino limitado e validação diferentes, e no final dos testes, para obter o modelo mais preciso é feito a média das avaliações de cada modelo.

- Assim mitigamos o acaso da equação e podemos ter conjuntos de validações menores, o ponto negativo desse método é o tempo, quanto mais dobras maior será o tempo.

### Incompatibilidade dos dados: 

- Por fim, devemos ressaltar uma informação, se os conjuntos de teste e validação não forem representativos dos dados obtidos em produção, então o seu modelo não terá um bom desempenho em produção, caindo naquele mesmo problema da representatividade dos dados discutido anteriormente.

### Extra: Teorema No free lunch:

- Esse teorema fala que a priori não há distinção entre algoritmos de aprendizado, isto é, se não fizermos suposições a partir dos dados, para sabermos qual o melhor modelo, teríamos que treinar, testar e avaliar todos, sendo uma tarefa praticamente impossível.

- Logo podemos fazer suposições com os dados e dessa maneira escolher modelos razoáveis para aquela situação
